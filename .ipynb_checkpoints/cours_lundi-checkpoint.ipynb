```python
import numpy as np

def f1(X):
    X_squared = X ** 2  
    dot_product = np.dot(X, X)  
    return X_squared * dot_product
```

```python
def f2(X, a_prime):

    return X + a_prime
```

```python
import numpy as np


X = np.array([1, 2, 3])

a_prime = 5


def f1(X):
    return 2 * X + 4  

def f2(X, a):
    return X + a  


result_f1 = f1(X)
result_f2 = f2(X, a_prime)

print("Résultat de f1(X) :", result_f1)
print("Résultat de f2(X, a_prime) :", result_f2)

```
